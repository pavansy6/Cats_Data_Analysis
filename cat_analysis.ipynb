```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"cats_dataset.csv")
```

```python
breed_counts = df['Breed'].value_counts()
sns.barplot(x=breed_counts.index, y=breed_counts.values)
plt.title('Breed Distribution')
plt.xlabel('Breed')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
color_counts = df['Color'].value_counts()
sns.barplot(x=color_counts.index, y=color_counts.values)
plt.title('Color Distribution')
plt.xlabel('Color')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['Age (Years)'], bins = 20, kde = True, color = 'black')
plt.title('Age Distribution')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['Weight (kg)'], bins = 10, kde = True, color = 'black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

# Bivariate Analysis

```python
correlation = df['Age (Years)'].corr(df['Weight (kg)'])
print("Correlation between Age and Weight: ",correlation)
```

```python
sns.boxplot(x='Breed', y='Weight (kg)', data=df)
plt.title('Weight Distribution by Breed')
plt.xlabel('Breed')
plt.ylabel('Weight (kg)')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x='Breed', y='Age (Years)', data=df)
plt.title('Age Distribution by Breed')
plt.xlabel('Breed')
plt.ylabel('Age (Years)')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x='Gender', y='Weight (kg)', data=df)
plt.title('Weight Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
sns.boxplot(x='Gender', y='Age (Years)', data=df)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age (Years)')
plt.show()
```
